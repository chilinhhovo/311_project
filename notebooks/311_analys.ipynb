```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

complaints = pd.read_csv("data/lift_complaints.csv")
complaints = complaints.dropna(subset=["Latitude", "Longitude"])

complaints_gdf = gpd.GeoDataFrame(
    complaints,
    geometry=gpd.points_from_xy(complaints["Longitude"], complaints["Latitude"]),
    crs="EPSG:4326"
)

# Spatial join: only complaints inside NYCHA developments
joined = gpd.sjoin(complaints_gdf, nycha_gdf, how="inner", predicate="within")

# Save to CSV
joined.to_csv("data/complaints_in_nycha.csv", index=False)
```
